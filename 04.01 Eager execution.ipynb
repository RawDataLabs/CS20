# 04.01 Eager execution

Graph 상황에서 Debug가 어려운 부분을 유연한 플랫폼을 만들기 위해 Eager execution을 추가 하였다.

###  실제 실험

```python
import tensorflow as tf # version >= 1.50
import tensorflow.contrib.eager as tfe
tfe.enable_eager_execution()
```

pdb.set_trace() # python debugging 툴과 호환이 가능하며  
아주 쉽게 Debugging 기능을 제공한다.

```python
i = tf.constant(0)
while i < 1000:
    i = tf.add(i, 1)
    print("I could do this all day! %d" % i)
```

# Eager execution 사용하는 이유

### 유연한 프레임 워크를 원한다면  
    - 파이썬 제어 흐름 및 데이터 구조로 실험 가능 하다.  
### 새로운 모델 개발  
    - 즉각적인 오류보고는 디버깅을 할수 있어 단순화 됩니다.
### TensorFlow의 새로운 기능을 사용하고 싶다면
    - 실행을 통해 Python REPL에서 TF API를 탐색 할 수 있다.

### Tensor Flow 1.5 이상 부터 사용 가능하다.
